```python
train_dataset=Dataset.from_dict({
    'text':train_text,
    'label':train_label
})

test_dataset=Dataset.from_dict({
    'text':eval_text,
    'label':eval_label
})
```

```python
from transformers import AutoTokenizer, AutoModelForMaskedLM
```

```python
model_name='distilbert-base-uncased'
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
def tokenization_fun(data):
  return tokenizer(data['text'],padding='max_length',truncation=True,max_length=128)
```

```python
tokenized_data=raw_dataset.map(tokenization_fun)
```

```python
tokenized_data=tokenized_data.remove_columns(['text'])
```

```python
train_dataset_trainer=tokenized_data['train']
test_dataset_trainer=tokenized_data['test']
```

```python
# 80
```

```python
from transformers import AutoModelForSequenceClassification
```

```python
model=AutoModelForSequenceClassification.from_pretrained(model_name,num_labels=2)
```

```python
from transformers import TrainingArguments
```

```python
train_args=TrainingArguments(
    output_dir="./model_result",
    num_train_epochs=3,
    eval_strategy='epoch',
    save_strategy='epoch',
    load_best_model_at_end=True

)
```

```python
0,1--> 0-> 0.4, 1--> 0.6--->1
```

```python
# -1
```

```python
def compute(pred):
  label=pred.label_ids   #[0,1]
  preds=pred.predictions.argmax(-1)   #[0-0.3,1-0.7]

  acc=accuracy_score(label,preds)
  return {'accuracy':acc}
```

```python
from transformers import Trainer
```

```python
test_dataset_trainer
```

```python
trainer=Trainer(
    model=model,
    args=train_args,
    train_dataset=train_dataset_trainer,
    eval_dataset=test_dataset_trainer,
    compute_metrics=compute,
    tokenizer=tokenizer,

)
```

```python
print('Staring the traing preocess')
try:
  trainer.train()
  print('model has been tuned sucessfully')
except Exception as e:
  print('the model tuning was unsucessful',e)
```

```python
SAVE_PATH="./fintuned_model"
```

```python
try:
  trainer.save_model(SAVE_PATH)
  tokenizer.save_pretrained(SAVE_PATH)
  print('Model saved sucessfully')
except Exception as e:
  print("Model saving failed",e)
```

```python
fintuned_token=AutoTokenizer.from_pretrained(SAVE_PATH)
```

```python
fintuned_model=AutoModelForSequenceClassification.from_pretrained(SAVE_PATH)
```
